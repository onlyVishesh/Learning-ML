```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x,y):
    m_curr = c_curr = 0
    rate = 0.01
    n=len(x)
    plt.scatter(x,y)
    for i in range(pow(10,4)):
        y_pred = m_curr*x+c_curr
        plt.plot(x,y_pred)
        md = -(2/n)*sum(x*(y-y_pred))
        cd = -(2/n)*sum(y-y_pred)
        m_curr -= rate*md
        c_curr -= rate*cd
```

```python
class gradient_descent1:
    def __init__(self):
        self.m_curr = self.c_curr = 0
    def fit(self,x,y):
        rate = 0.01
        n=len(x)
        plt.scatter(x,y)
        for i in range(pow(10,4)):
            y_pred = self.m_curr*x+self.c_curr
            plt.plot(x,y_pred)
            md = -(2/n)*sum(x*(y-y_pred))
            cd = -(2/n)*sum(y-y_pred)
            self.m_curr -= rate*md
            self.c_curr -= rate*cd
```

```python
x=np.array([1,2,3,4,5])
y=np.array([3,5,7,11,13])
```

```python
gradient_descent(x,y)
```

```python
gr = gradient_descent1()
```

```python
gr.fit(x,y)
```

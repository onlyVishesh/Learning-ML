# Matplotlib

```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# linspace is a numpy function for making an array of evenly seperated values
p = np.linspace(0,10,100) #np.arange(0,11,0.1)
q = np.sin(p)
r = np.cos(p)
s = np.log(p)
```

```python
# Plot the chart
plt.plot(p, q)  #graph of sinx
plt.plot(p,s)   #graph of logx
# to show graph
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(p,q,"r--")
plt.subplot(2,2,2)
plt.plot(p,r,"b-*")
plt.subplot(2,2,3)
plt.plot(p,s,"c*")
plt.subplot(2,2,4)
plt.plot(p,q,"g-+")
```

```python
plt.plot(p,q)
plt.xlabel("Angles")
plt.ylabel("Values")
plt.title("Sine wave")
plt.show()
```

```python
x = np.linspace(1,40,150)
y = 9*x+5
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Linear graph")
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-10,10,50) #Higher the value of division smoother the curve
y = 8*(x**2)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Parabola")
plt.plot(x,y,"r*") # We can shife to origin by adding constant to x and y 
plt.plot(x+10,y,"g+")
plt.show()
```

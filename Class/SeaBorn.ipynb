```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# taking inbuild dataset
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="time")
```

```python
#lineplot
sns.lineplot(data=tips,x="total_bill",y="tip",hue="time")
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.kdplot(data=tips,x="total_bill",y="tip",hue="sex")
```

```python
from sklearn import preprocessing
labal_encode = preprocessing.LabalEncoder()
titanic["alive"] = labal_encoder.fit_transform(titanic["alive"])
```

```python
sns.kdplot(data=titanic,x="alive",hue="sex")
```
